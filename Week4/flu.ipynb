```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_flu = pd.read_csv('flu.csv')
display(df_flu.head())
print(df_flu.shape)

print("Data Shape:", df_flu.shape)
print("\nData Info:")
df_flu.info()

missing_values = df_flu.isnull().sum()
missing_percentage = (missing_values / len(df_flu)) * 100
print("\nMissing Values:")
print(missing_values)
print("\nMissing Value Percentage:")
print(missing_percentage)

numerical_features = df_flu.select_dtypes(include=['number'])
print("\nDescriptive Statistics for Numerical Features:")
display(numerical_features.describe())


target_variable = 'Diagnosis'
print(f"\nTarget Variable: {target_variable}")
print(f"Target Variable Data Type: {df_flu[target_variable].dtype}")
print(f"Target Variable Distribution:\n{df_flu[target_variable].value_counts(normalize=True)}")

predictor_variables = df_flu.columns.drop(target_variable)
print("\nPotential Predictor Variables:")
print(predictor_variables)
for col in predictor_variables:
    if df_flu[col].dtype == 'object':
        print(f"\n{col} Distribution:\n{df_flu[col].value_counts(normalize=True)}")
    else:
        print(f"\n{col} Summary:\n{df_flu[col].describe()}")

print("\nInitial Observations:")

for col in df_flu.columns:
    if df_flu[col].dtype == 'float64':
        df_flu[col + '_imputed'] = df_flu[col].isnull()
        df_flu[col] = df_flu[col].fillna(df_flu[col].median())
    elif df_flu[col].dtype == 'object':
        df_flu[col + '_imputed'] = df_flu[col].isnull()
        df_flu[col] = df_flu[col].fillna(df_flu[col].mode()[0])

for col in df_flu.select_dtypes(include=['float64']):
    if col not in ['Age', 'Temperature', 'DurationOfIllness']:
        Q1 = df_flu[col].quantile(0.25)
        Q3 = df_flu[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_flu[col] = np.clip(df_flu[col], lower_bound, upper_bound)

df_flu['InitialPCRDiagnosis'] = df_flu['InitialPCRDiagnosis'].astype(str)

df_flu['Age'] = df_flu['Age'].replace(0, df_flu['Age'].median())

df_flu.drop_duplicates(inplace=True)

df_cleaned = df_flu.copy()
display(df_cleaned.head())

df_engineered = df_cleaned.copy()

for col in ['Fever', 'Coughing', 'ShortnessOfBreath', 'GroundGlassOpacity', 'Diarrhea', 'SoreThroat', 'NauseaVomitting', 'Fatigue']:
    df_engineered[col] = df_engineered[col].map({'Yes': 1, 'No': 0}).fillna(0).astype(int)

df_engineered['Fever_Cough'] = df_engineered['Fever'] * df_engineered['Coughing']
df_engineered['Combined_Respiratory'] = df_engineered['Coughing'] + df_engineered['ShortnessOfBreath'] + df_engineered['GroundGlassOpacity']


numerical_cols = ['Age', 'Temperature', 'DurationOfIllness', 'neutrophil', 'serumLevelsOfWhiteBloodCell', 'lymphocytes', 'CReactiveProteinLevels']
for col in numerical_cols:
    df_engineered[col] = pd.to_numeric(df_engineered[col], errors='coerce')
    df_engineered[col + '_squared'] = df_engineered[col]**2


categorical_cols = ['Sex', 'InitialPCRDiagnosis', 'CTscanResults', 'RiskFactors']

df_engineered = pd.get_dummies(df_engineered, columns=categorical_cols, drop_first=True)

display(df_engineered.head())

display(df_engineered.describe())

numerical_cols = ['Age', 'Temperature', 'DurationOfIllness', 'neutrophil', 'serumLevelsOfWhiteBloodCell', 'lymphocytes', 'CReactiveProteinLevels']
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df_engineered[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()


plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(x=df_engineered[col])
    plt.title(col)
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(x='Diagnosis', y=col, data=df_engineered)
    plt.title(f'{col} vs Diagnosis')
plt.tight_layout()
plt.show()

corr_matrix = df_engineered[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

categorical_cols = [col for col in df_engineered.columns if col.startswith('Sex_') or col.startswith('InitialPCRDiagnosis_') or col.startswith('CTscanResults_') or col.startswith('RiskFactors_')]
plt.figure(figsize=(15, 30))
for i, col in enumerate(categorical_cols):
    plt.subplot(len(categorical_cols) // 3 + 1, 3, i + 1)
    sns.countplot(x=col, hue='Diagnosis', data=df_engineered)
    plt.title(col)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Observations:")
```

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, classification_report


df_encoded = pd.get_dummies(df_flu, drop_first=True)

target = 'Diagnosis_H1N1'
if target not in df_encoded.columns:
    raise ValueError(f"Target '{target}' tidak ditemukan")

df_clean = df_encoded.dropna(subset=[target])
X = df_clean.drop(columns=target)
y = df_clean[target]

imputer = SimpleImputer(strategy='mean')
X_filled = imputer.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_filled)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
y_prob = clf.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))

fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f'ROC AUC = {roc_auc_score(y_test, y_prob):.2f}', linewidth=2)
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()


print(f"Accuracy : {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision : {precision_score(y_test, y_pred):.4f}")
print(f"Recall : {recall_score(y_test, y_pred):.4f}")
print(f"F1-Score : {f1_score(y_test, y_pred):.4f}")
print(f"AUC : {roc_auc_score(y_test, y_prob):.4f}")

print("\nRumus Matematis dan Penjelasan:")
print("Accuracy: (TP + TN) / (TP + TN + FP + FN)")
print("  - TP (True Positif): Prediksi benar kelas positif, TN (True Negatif): Prediksi benar kelas negatif, FP (False Positif): Prediksi salah kelas positif, FN (False Negatif): Prediksi salah kelas negatif")
print("Precision: TP / (TP + FP)")
print("  - Mengukur proporsi kasus positif yang diprediksi dengan benar dari seluruh prediksi positif.")
print("Recall (Sensitivity): TP / (TP + FN)")
print("  - Mengukur proporsi kasus positif yang diprediksi dengan benar dari seluruh kasus positif aktual.")
print("F1-score: 2 * (Presisi * Recall) / (Presisi + Recall)")
print("  - Rata-rata harmonik dari presisi dan recall, menyeimbangkan kedua metrik.")
print("AUC (Area Under the Curve): Area di bawah kurva ROC.")
print("  - Menggambarkan kemampuan model dalam membedakan antar kelas.")
```
