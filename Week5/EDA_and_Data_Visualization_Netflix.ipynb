```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('netflix_titles.csv')
display(df.head())


print("Data Shape:", df.shape)
display(df.info())

display(df.describe(include='all'))
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
print("\nMissing Values Percentage:\n", missing_percentage)

plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# Data Distribution Visualization
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
df['type'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Distribution of Content Type')

plt.subplot(1, 3, 2)
df['rating'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribution of Ratings')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 3, 3)
sns.histplot(df['release_year'], kde=True, color='orange')
plt.title('Distribution of Release Year')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
top_countries = df['country'].value_counts().nlargest(10)
top_countries.plot(kind='bar', color='purple')
plt.title('Top 10 Countries')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
top_listed_in = df['listed_in'].value_counts().nlargest(10)
top_listed_in.plot(kind='bar', color='pink')
plt.title('Top 10 Listed In')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Cleaning
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df.dropna(subset=['date_added'], inplace=True)
print("\nInconsistencies in 'release_year':")
print(df[df['release_year'] < 1900])

duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
df.drop_duplicates(inplace=True)
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed', errors='coerce')
df.dropna(subset=['date_added'], inplace=True) # Drop rows with invalid date format



# Feature engineering
df['year_added'] = pd.to_datetime(df['date_added']).dt.year
df['year_added'] = df['year_added'].astype(int)
df['is_movie'] = df['type'].apply(lambda x: 1 if x == 'Movie' else 0)

top_genres = df['listed_in'].str.split(',').explode().str.strip().value_counts().nlargest(10).index
for genre in top_genres:
    df[f'is_{genre.lower().replace(" ", "_")}'] = df['listed_in'].str.contains(genre, case=False, regex=False).astype(int)

# Document new features (included as comments within the code)
# 'year_added': Year extracted from 'date_added' column.  Represents the year when the show was added to Netflix.
# 'is_movie': Binary indicator (1 for Movie, 0 for TV Show).  Helps distinguish between movie and TV show entries.
# 'is_International TV Shows', 'is_Dramas', 'is_Comedies', etc.: Binary indicators (1 if the genre is present in 'listed_in', 0 otherwise).  Allows for easy filtering and analysis based on genre.  The specific genres were selected based on the top 10 most frequent genres in the 'listed_in' column.

display(df.head())

plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['year_added'], kde=True, color='skyblue')
plt.title('Distribution of Year Added')

plt.subplot(1, 2, 2)
genre_cols = [col for col in df.columns if col.startswith('is_')]
genre_counts = df[genre_cols].sum()
genre_counts.plot(kind='bar', color='lightcoral')
plt.title('Distribution of Genres')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
df['is_movie'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Movies vs. TV Shows')
plt.xticks([0, 1], ['TV Shows', 'Movies'])
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='year_added', y='release_year', data=df, color='purple', alpha=0.5)
plt.title('Year Added vs. Release Year')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='is_movie', y='duration', data=df, palette=['skyblue', 'lightcoral'])
plt.title('Duration of Movies vs. TV Shows')
plt.xticks([0, 1], ['TV Shows', 'Movies'])
plt.show()

top_genres = df[genre_cols].sum().nlargest(3).index
plt.figure(figsize=(10, 6))
for i, genre in enumerate(top_genres):
    sns.histplot(df[df[genre] == 1]['rating'], label=genre, kde=False, color=plt.cm.get_cmap('tab10')(i), alpha=0.7, element='step')
plt.title('Distribution of Ratings across Top 3 Genres')
plt.legend()
plt.show()

numerical_features = ['year_added', 'is_movie'] + list(genre_cols)
plt.figure(figsize=(12, 10))
sns.heatmap(df[numerical_features].corr(), annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

print("Penjelasan Matematis dan Fungsi-Fungsi yang Digunakan:")
print("\n1. Deskripsi Statistik (df.describe()):")
print("  - Memberikan ringkasan statistik deskriptif dari data, termasuk mean, standar deviasi, nilai minimum, kuartil, dan nilai maksimum untuk kolom numerik. Ini membantu memahami distribusi data dan mengidentifikasi outlier potensial.")
print("  - Fungsi ini tidak menggunakan persamaan matematika yang spesifik, tetapi mengandalkan perhitungan statistik dasar.")


print("\n2. Persentase Nilai Hilang (missing_percentage):")
print("  - Menghitung persentase nilai yang hilang di setiap kolom menggunakan rumus:")
print("    Persentase = (Jumlah Nilai Hilang / Total Jumlah Data) * 100")
print("  - Ini membantu mengukur kualitas data dan mengidentifikasi area yang perlu dibersihkan.")

print("\n3. Visualisasi Distribusi Data (Histogram, Bar Chart):")
print("  - Histogram: Menampilkan distribusi frekuensi dari variabel numerik (misalnya, 'release_year').  ")
print("  - Bar Chart: Menampilkan frekuensi dari variabel kategorik (misalnya, 'type', 'rating', 'country', 'listed_in').")
print("  - Tidak ada persamaan matematika yang digunakan secara eksplisit, tetapi fungsi ini mengandalkan representasi visual dari data.")

print("\n4. Feature Engineering:")
print("  - 'year_added': Tahun diekstrak dari kolom 'date_added' menggunakan fungsi datetime.dt.year. Tidak ada persamaan matematika yang digunakan.")
print("  - 'is_movie': Membentuk kolom baru menggunakan fungsi lambda: 1 jika 'type' adalah 'Movie', 0 jika tidak. Ini adalah fungsi logis sederhana.")

print("\n5. Genre Feature Engineering:")
print("  - Menghasilkan kolom baru (is_International TV Shows, is_Dramas, etc.) berdasarkan keberadaan genre tertentu dalam kolom 'listed_in' ")
print("  - str.contains(): Fungsi ini melakukan pengecekan kesamaan string. Nilai 1 diberikan jika suatu genre ditemukan dan 0 jika tidak. Tidak melibatkan persamaan matematika rumit. Ini adalah operasi boolean.")


print("\n6. Korelasi (df[numerical_features].corr()):")
print("  - Menghitung koefisien korelasi antara fitur numerik untuk mengukur kekuatan dan arah hubungan linear antara variabel. Koefisien korelasi antara dua variabel x dan y diberikan oleh rumus:")
print("   Korelasi (r) = Cov(x,y) / (std(x) * std(y))")
print("     dimana:")
print("     - Cov(x,y) adalah kovariansi antara x dan y.")
print("     - std(x) dan std(y) adalah standar deviasi x dan y.")
print("   Nilai korelasi berkisar dari -1 hingga 1. Nilai -1 menunjukkan korelasi negatif yang sempurna, nilai 1 menunjukkan korelasi positif sempurna, dan nilai 0 menunjukkan tidak ada korelasi linier.")
```
